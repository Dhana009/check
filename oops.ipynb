Class: A class can be thought of as a blueprint for creating an object. For example, a class "Car" can have attributes like "color", "make", and "model" and methods like "drive()" and "stop()". When you create an object of class "Car", you can set the values of its attributes and call its methods.

Object: An object is a specific instance created from a class. For example, you can create an object of class "Car" called "myCar" and set its color to "red", make to "Toyota", and model to "Camry". You can then call the drive() and stop() methods on "myCar".

Inheritance: One class can inherit properties and methods from another class. For example, you can have a class "ElectricCar" that inherits from the "Car" class. This means that "ElectricCar" has all the attributes and methods of "Car" and can also have additional attributes and methods specific to electric cars.

Polymorphism: Objects of different classes can be treated as objects of a common superclass. For example, a function that accepts an object of the superclass "Vehicle" can accept an object of the class "Car" or "ElectricCar" or "Bike" and can perform the same action on all of them.

Abstraction: Hiding complexity and only showing the necessary information. For example, A music player application, we interact with the buttons like play, pause, next, previous, and volume but we don't need to know how the application works internally, how it read the mp3 file, how it decodes the mp3 format, how it sends the signal to speaker, we only need to know the result of our interaction.

Encapsulation: Keeping the internal workings of an object hidden and only allowing controlled access. For example, a Bank account class, we can access the account balance, deposit and withdraw money but we don't have access to the internal workings of the class, like how the balance is stored, how the transactions are processed, how the security measures are implemented.

```python
#implementing method overloading

class Quantity:
    def __init__(self,v=0):
        self._value=v
    
    def __add__(self,other):
        new_value=self._value+other._value
        return Quantity(new_value)
    
    def __sub__(self,other):
        new_value=self._value - other._value
        return Quantity(new_value) 
    
    def __mul__(self,other):
        new_value=self._value * other._value
        return Quantity(new_value) 
    
    def __pow__(self,other):
        new_value=self._value ** other._value
        return Quantity(new_value)
    
    #division
    def __truediv__(self,other):
        new_value=self._value / other._value
        return Quantity(new_value) 
    
    #floor division
    def __floordiv__(self,other):
        new_value=self._value // other._value
        return Quantity(new_value)
    
    #remainder
    def __mod__(self,other):
        new_value=self._value % other._value
        return Quantity(new_value) 
    
    #lessthan
    def __lt__(self,other):
        new_value=self._value < other._value
        return Quantity(new_value) 
   
    #lessthan or equal to
    def __le__(self,other):
        new_value=self._value <= other._value
        return Quantity(new_value)
    
    #equal to
    def __eq__(self,other):
        new_value=self._value == other._value
        return Quantity(new_value)
    
    #not equal to
    def __ne__(self,other):
        new_value=self._value != other._value
        return Quantity(new_value)
    
    #greter than
    def __gt__(self,other):
        new_value=self._value > other._value
        return Quantity(new_value)
    
    #greter than or equal to
    def __ge__(self,other):
        new_value=self._value >= other._value
        return Quantity(new_value)
    
    def __str__(self):
        return str(self._value)
        
```

```python
#Sure, here's an example in Python that demonstrates polymorphism:
class Shape:
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def calculate_area(self):
        return 3.14 * (self.radius ** 2)

class Square(Shape):
    def __init__(self, side):
        self.side = side
        
    def calculate_area(self):
        return self.side ** 2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

def print_area(shape):
    print(shape.calculate_area())

circle = Circle(5)
square = Square(6)
triangle = Triangle(3,4)

print_area(circle) # prints 78.5
print_area(square) # prints 36
print_area(triangle) # prints 6.0

triangle.calculate_area()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance -= amount
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

class SavingAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.__interest_rate = interest_rate
        
    def add_interest(self):
        interest = self.__interest_rate * self.get_balance()
        self.deposit(interest)

    def get_interest_rate(self):
        return self.__interest_rate
    
class CurrentAccount(BankAccount):
    def __init__(self, balance, overdraft_limit):
        super().__init__(balance)
        self.__overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if self.__balance - amount >= -self.__overdraft_limit:
            self.__balance -= amount
        else:
            print("Insufficient funds, Overdraft limit reached")
    def get_overdraft_limit(self):
        return self.__overdraft_limit

def account_summary(account):
    if isinstance(account, SavingAccount):
        print("Account Type: Saving Account")
        print("Balance: ", account.get_balance())
        print("Interest Rate: ", account.get_interest_rate())
    elif isinstance(account, CurrentAccount):
        print("Account Type: Current Account")
        print("Balance: ", account.get_balance())
        print("Overdraft Limit: ", account.get_overdraft_limit())

saving_account = SavingAccount(5000, 0.03)
current_account = CurrentAccount(3000, -1000)
account_summary(saving_account)
account_summary(current_account)

```

```python
import logging
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
        logging.basicConfig(filename='bank.log', level=logging.INFO)

    def deposit(self, amount):
        try:
            if amount <= 0:
                raise ValueError("Invalid deposit amount")
            self.__balance += amount
            logging.info(f"Deposited {amount} to account. New balance: {self.__balance}")
        except ValueError as e:
            logging.error(e)
            print(e)

    def withdraw(self, amount):
        try:
            if self.__balance < amount:
                raise ValueError("Insufficient funds")
            self.__balance -= amount
            logging.info(f"Withdrew {amount} from account. New balance: {self.__balance}")
        except ValueError as e:
            logging.error(e)
            print(e)
    
    def get_balance(self):
        return self.__balance

class SavingsAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.__interest_rate = interest_rate

    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        logging.info(f"Added interest of {interest} to account. New balance: {self.get_balance()}")

savings_account = SavingsAccount(1000, 0.03)
savings_account.add_interest()
print(savings_account.get_balance())
savings_account.withdraw(100)
print(savings_account.get_balance())

```

```python
import logging

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
        self.logger = logging.getLogger(__name__)

    def deposit(self, amount):
        try:
            if amount <= 0:
                raise ValueError("Invalid deposit amount")
            self.__balance += amount
            self.logger.info(f"Deposited {amount}, new balance is {self.__balance}")
        except ValueError as e:
            self.logger.exception(e)

    def withdraw(self, amount):
        try:
            if amount <= 0:
                raise ValueError("Invalid withdraw amount")
            if self.__balance < amount:
                raise ValueError("Insufficient funds")
            self.__balance -= amount
            self.logger.info(f"Withdrew {amount}, new balance is {self.__balance}")
        except ValueError as e:
            self.logger.exception(e)

    def get_balance(self):
        return self.__balance
    
    def check_overdraft(self):
        if self.__balance < 0:
            self.logger.warning("Account is in overdraft")
            return True
        return False
    
class SavingsAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.__interest_rate = interest_rate
        
    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        self.logger.info(f"Added interest of {interest}, new balance is {self.get_balance()}")


logging.basicConfig(level=logging.INFO)
account = SavingsAccount(1000, 0.05)
account.add_interest()
print(account.get_balance()) # prints 1050.0
account.withdraw(500)
print(account.get_balance()) # prints 550.0
account.withdraw(1000) # raises an exception and logs "Insufficient funds"

try:
    account.withdraw(1000)
except ValueError as e:
    print(e)

if account.check_overdraft():
    print("Account is in overdraft")
else:
    print("Account is not in overdraft")

```

```python
import logging

class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary
        self.logger = logging.getLogger(__name__)

    def get_name(self):
        return self.__name

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        try:
            if salary < 0:
                raise ValueError("Invalid salary")
            self.__salary = salary
            self.logger.info(f"Salary for {self.__name} updated to {self.__salary}")
        except ValueError as e:
            self.logger.exception(e)

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.__department = department
        self.__employees = []

    def get_department(self):
        return self.__department

    def add_employee(self, employee):
        self.__employees.append(employee)
        self.logger.info(f"{employee.get_name()} added to {self.__department} department")

    def remove_employee(self, employee):
        self.__employees.remove(employee)
        self.logger.info(f"{employee.get_name()} removed from {self.__department} department")

class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.__programming_language = programming_language

    def get_programming_language(self):
        return self.__programming_language

class Designer(Employee):
    def __init__(self, name, salary, design_tool):
        super().__init__(name, salary)
        self.__design_tool = design_tool

    def get_design_tool(self):
        return self.__design_tool


logging.basicConfig(level=logging.INFO,filename='application.log')


manager = Manager("John Doe", 80000, "IT")

dev1 = Developer("Jane Smith", 60000, "Python")
dev2 = Developer("Bob Johnson", 65000, "JavaScript")
designer = Designer("Emma Williams", 70000, "Photoshop")

manager.add_employee(dev1)
manager.add_employee(dev2)
manager.add_employee(designer)

manager.remove_employee(dev1)

#dev2.set_salary(-10000)

try:
    dev2.set_salary(10000)
except ValueError as e:
    print(e)
```

```python
import logging

class Aerofield:
    def __init__(self, name, location):
        self.__name = name
        self.__location = location
        self.__runways = []
        self.logger = logging.getLogger(__name__)
        
    def add_runway(self, runway):
        self.__runways.append(runway)
        self.logger.info(f"Added runway {runway.get_name()} to {self.__name} aerofield")
        
    def get_runways(self):
        return self.__runways
    
    def get_name(self):
        return self.__name
    
    def get_location(self):
        return self.__location

class Runway:
    def __init__(self, name, length, width):
        self.__name = name
        self.__length = length
        self.__width = width
        self.__aircrafts = []
        self.logger = logging.getLogger(__name__)
        
    def add_aircraft(self, aircraft):
        if self.is_space_available():
            self.__aircrafts.append(aircraft)
            self.logger.info(f"Added aircraft {aircraft.get_name()} to runway {self.__name}")
        else:
            self.logger.warning(f"No space available on runway {self.__name}")
            
    def remove_aircraft(self, aircraft):
        self.__aircrafts.remove(aircraft)
        self.logger.info(f"Removed aircraft {aircraft.get_name()} from runway {self.__name}")
        
    def is_space_available(self):
        if len(self.__aircrafts) < 5:
            return True
        return False
    
    def get_name(self):
        return self.__name
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
class Aircraft:
    def __init__(self, name, manufacturer, year):
        self.__name = name
        self.__manufacturer = manufacturer
        self.__year = year
        
    def get_name(self):
        return self.__name
    
    def get_manufacturer(self):
        return self.__manufacturer
    
    def get_year(self):
        return self.__year

logging.basicConfig(level=logging.INFO,filename='airfield.log')

jfk = Aerofield("John F Kennedy International Airport", "New York")
r1 = Runway("R1", 10000, 200)
r2 = Runway("R2", 9000, 150)

jfk.add_runway(r1)
jfk.add_runway(r2)

b737 = Aircraft("Boeing 737", "Boeing", 2020)
b747 = Aircraft("Boeing 747", "Boeing", 2019)
a320 = Aircraft("Airbus A320", "Airbus", 2018)

r1.add_aircraft(b737)
r1.add_aircraft(b747)
r1.add_aircraft(a320)

r2.add_aircraft(b737)
r2.add_aircraft(b747)
r2.add_aircraft(a320)

print(f"Runways at {jfk.get_name()} aerofield:")
for runway in jfk.get_runways():
    print(runway.get_name())
    print("Aircrafts on runway:")
    for aircraft in runway._Runway__aircrafts:
        print(aircraft.get_name())
```
