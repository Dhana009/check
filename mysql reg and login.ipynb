```python
from tkinter import *
from PIL import ImageTk
from tkinter.ttk import Combobox

root=Tk()
root.geometry('1300x750+10+10')
root.title('Registration Form')

bgimage=PhotoImage(file='loginbg.png')
bglabel = Label(root,image=bgimage)

bglabel.place(x=0,y=0)

registerframe = Frame(root,width=600,height=600)
registerframe.place(x=630,y=30)

titlelabel = Label(registerframe,text='Registration form',font=('Roboto',12,'bold'),bg='grey',
                   fg='white')
titlelabel.place(x=20,y=5)

firstname = Label(registerframe,text="First Name",font=('arial',15,'bold'),fg='blue')
firstname.place(x=30,y=65)

firstentry=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
firstentry.place(x=30,y=95)

lastname = Label(registerframe,text="Last Name",font=('arial',15,'bold'),fg='blue')
lastname.place(x=350,y=65)

lastentry=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
lastentry.place(x=350,y=95)

contact = Label(registerframe,text="Contact Number",font=('arial',15,'bold'),fg='blue')
contact.place(x=30,y=180)

contact=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
contact.place(x=30,y=210)

email = Label(registerframe,text="Email",font=('arial',15,'bold'),fg='blue')
email.place(x=350,y=180)

email=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
email.place(x=350,y=210)


comboquestion=Combobox(registerframe,font=('arial',15,'bold'),state='readonly')
comboquestion['values']=('select','what is your name?','what is your pet name?',
                         'what is your date of birth? ')

comboquestion.place(x=30,y=320)
comboquestion.current(0)

security = Label(registerframe,text="Security",font=('arial',15,'bold'),fg='blue')
security.place(x=30,y=290)

#security=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
#security.place(x=30,y=320)

answer = Label(registerframe,text="Answer",font=('arial',15,'bold'),fg='blue')
answer.place(x=350,y=290)

answer=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
answer.place(x=350,y=320)

password = Label(registerframe,text="Password",font=('arial',15,'bold'),fg='blue')
password.place(x=30,y=400)

password=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
password.place(x=30,y=430)

confirmpassword = Label(registerframe,text="Confirm Password",font=('arial',15,'bold'),fg='blue')
confirmpassword.place(x=350,y=400)

confirmpassword=Entry(registerframe,font=('times new roman',18),bg='lightgray',fg='black',bd=2,width=18)
confirmpassword.place(x=350,y=430)

agree=Checkbutton(registerframe,text='I agree to terms and conditions',font=('times new roman',18,'bold'),
                 onvalue=1,offvalue=0,fg='black')
agree.place(x=30,y=490)

buttonimage=PhotoImage(file='button.png')
registerbutton=Button(registerframe,image=buttonimage,bd=0,cursor='hand2')
registerbutton.place(x=240,y=530)

loginimage=PhotoImage(file='login.png')
loginbutton=Button(root,image=loginimage,bd=1,bg='gray',cursor='hand2')
loginbutton.place(x=250,y=300)

root.mainloop()
```
