```python
minutes=list(range(1,9))
player1=[1,2,3,4,4,4,6,6]
player2=[1,3,3,3,4,5,5,5]
player3=[1,1,1,2,2,2,4,4]
labels=['player1','player2','player3']
colors=['red','black','green']
plt.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)
plt.title('My awesome stack plot')
plt.legend()
plt.show()
```

```python
fig,(ax,ax1)=plt.subplots(nrows=2,ncols=1,sharex=True)
ax.plot(tips.total_bill,tips.tip,color='red',label='all tips')
ax1.plot(tips.total_bill,tips.size,color='green',label='all sizes')
ax.legend(loc=(0.07,0.05))
ax.xlabel('total_bill')
ax1.legend(loc=(0.07,0.05))
ax1.xlabel('total_bill')
plt.show()
```

```python
fig1,ax=plt.subplots()
fig2,ax1=plt.subplots()
ax.plot(tips.total_bill,tips.tip,color='red',label='all tips')
ax1.plot(tips.total_bill,tips.size,color='green',label='all sizes')
ax.legend(loc=(0.07,0.05))
ax.set_xlabel('total_bill')
ax.set_title('total_bill vs tip')
ax1.legend(loc=(0.07,0.05))
ax1.set_title('total_bill vs size')
ax1.set_xlabel('total_bill')
plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')

```
