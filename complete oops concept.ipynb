Class: A class can be thought of as a blueprint for creating an object. For example, a class "Car" can have attributes like "color", "make", and "model" and methods like "drive()" and "stop()". When you create an object of class "Car", you can set the values of its attributes and call its methods.

Object: An object is a specific instance created from a class. For example, you can create an object of class "Car" called "myCar" and set its color to "red", make to "Toyota", and model to "Camry". You can then call the drive() and stop() methods on "myCar".

Inheritance: One class can inherit properties and methods from another class. For example, you can have a class "ElectricCar" that inherits from the "Car" class. This means that "ElectricCar" has all the attributes and methods of "Car" and can also have additional attributes and methods specific to electric cars.

Polymorphism: Objects of different classes can be treated as objects of a common superclass. For example, a function that accepts an object of the superclass "Vehicle" can accept an object of the class "Car" or "ElectricCar" or "Bike" and can perform the same action on all of them.

Abstraction: Hiding complexity and only showing the necessary information. For example, A music player application, we interact with the buttons like play, pause, next, previous, and volume but we don't need to know how the application works internally, how it read the mp3 file, how it decodes the mp3 format, how it sends the signal to speaker, we only need to know the result of our interaction.

Encapsulation: Keeping the internal workings of an object hidden and only allowing controlled access. For example, a Bank account class, we can access the account balance, deposit and withdraw money but we don't have access to the internal workings of the class, like how the balance is stored, how the transactions are processed, how the security measures are implemented.

Object-oriented programming (OOP) is a programming paradigm that uses objects and their interactions to design applications and computer programs. The four main concepts of OOP are:

Encapsulation: The practice of keeping an object's internal state and behavior hidden from the outside world and providing a controlled interface for interacting with the object.

Inheritance: A way for one class to inherit properties and methods from another class.

Polymorphism: The ability for objects of different classes to be used interchangeably.

Abstraction: Hiding complexity and only showing the necessary information.

Here's an example in Python that demonstrates all four concepts of OOP:

```python
#implementing method overloading

class Quantity:
    def __init__(self,v=0):
        self._value=v
    
    def __add__(self,other):
        new_value=self._value+other._value
        return Quantity(new_value)
    
    def __sub__(self,other):
        new_value=self._value - other._value
        return Quantity(new_value) 
    
    def __mul__(self,other):
        new_value=self._value * other._value
        return Quantity(new_value) 
    
    def __pow__(self,other):
        new_value=self._value ** other._value
        return Quantity(new_value)
    
    #division
    def __truediv__(self,other):
        new_value=self._value / other._value
        return Quantity(new_value) 
    
    #floor division
    def __floordiv__(self,other):
        new_value=self._value // other._value
        return Quantity(new_value)
    
    #remainder
    def __mod__(self,other):
        new_value=self._value % other._value
        return Quantity(new_value) 
    
    #lessthan
    def __lt__(self,other):
        new_value=self._value < other._value
        return Quantity(new_value) 
   
    #lessthan or equal to
    def __le__(self,other):
        new_value=self._value <= other._value
        return Quantity(new_value)
    
    #equal to
    def __eq__(self,other):
        new_value=self._value == other._value
        return Quantity(new_value)
    
    #not equal to
    def __ne__(self,other):
        new_value=self._value != other._value
        return Quantity(new_value)
    
    #greter than
    def __gt__(self,other):
        new_value=self._value > other._value
        return Quantity(new_value)
    
    #greter than or equal to
    def __ge__(self,other):
        new_value=self._value >= other._value
        return Quantity(new_value)
    
    def __str__(self):
        return str(self._value)
  


q1=Quantity(10)
q2=Quantity(20)

q3=Quantity()
print(q1,q2)

q3=q1+q2
print(q3)

q3=q1-q2
print(q3)

q3=q1*q2
print(q3)

q3=q1**q2
print(q3)

q3=q1/q2
print(q3)

q3=q1//q2
print(q3)

q3=q1%q2
print(q3)
```

```python
#Sure, here's an example in Python that demonstrates polymorphism:
class Shape:
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def calculate_area(self):
        return 3.14 * (self.radius ** 2)

class Square(Shape):
    def __init__(self, side):
        self.side = side
        
    def calculate_area(self):
        return self.side ** 2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

def print_area(shape):
    print(shape.calculate_area())

circle = Circle(5)
square = Square(6)
triangle = Triangle(3,4)

print_area(circle) # prints 78.5
print_area(square) # prints 36
print_area(triangle) # prints 6.0

triangle.calculate_area()
```

```python
import logging

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
        self.logger = logging.getLogger(__name__)

    def deposit(self, amount):
        try:
            if amount <= 0:
                raise ValueError("Invalid deposit amount")
            self.__balance += amount
            self.logger.info(f"Deposited {amount}, new balance is {self.__balance}")
        except ValueError as e:
            self.logger.exception(e)

    def withdraw(self, amount):
        try:
            if amount <= 0:
                raise ValueError("Invalid withdraw amount")
            if self.__balance < amount:
                raise ValueError("Insufficient funds")
            self.__balance -= amount
            self.logger.info(f"Withdrew {amount}, new balance is {self.__balance}")
        except ValueError as e:
            self.logger.exception(e)

    def get_balance(self):
        return self.__balance
    
    def check_overdraft(self):
        if self.__balance < 0:
            self.logger.warning("Account is in overdraft")
            return True
        return False
    
class SavingsAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.__interest_rate = interest_rate
        
    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        self.logger.info(f"Added interest of {interest}, new balance is {self.get_balance()}")


logging.basicConfig(level=logging.INFO)
account = SavingsAccount(1000, 0.05)
account.add_interest()
print(account.get_balance()) # prints 1050.0
account.withdraw(500)
print(account.get_balance()) # prints 550.0
account.withdraw(1000) # raises an exception and logs "Insufficient funds"

try:
    account.withdraw(1000)
except ValueError as e:
    print(e)

if account.check_overdraft():
    print("Account is in overdraft")
else:
    print("Account is not in overdraft")

```

```python
import mysql.connector

class Database:
    
    def __init__(self, host, user, password, database):
        self.connection = mysql.connector.connect(
            host=host,
            user=user,
            password=password
        )
        self.cursor = self.connection.cursor()

        # Create the database if it doesn't exist
        self.cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
        self.connection.commit()

        # Use the database
        self.cursor.execute(f"USE {database}")
        self.connection.commit()
        

    def create_table(self, table_name, columns):
        query = f"CREATE TABLE if not exists {table_name}({columns})"
        self.cursor.execute(query)
        self.connection.commit()
        print(f"Table {table_name} has been created successfully.")

    def insert_data(self, table_name, values):
        query = f"INSERT INTO {table_name} VALUES({values})"
        self.cursor.execute(query)
        self.connection.commit()
        print(f"Data has been inserted into {table_name} table.")

    def select_data(self, table_name):
        query = f"SELECT * FROM {table_name}"
        self.cursor.execute(query)
        data = self.cursor.fetchall()
        print(data)

    def update_data(self, table_name, column, value, condition):
        query = f"UPDATE {table_name} SET {column} = {value} WHERE {condition}"
        self.cursor.execute(query)
        self.connection.commit()
        print(f"Data has been updated in {table_name} table.")

    def delete_data(self, table_name, condition):
        query = f"DELETE FROM {table_name} WHERE {condition}"
        self.cursor.execute(query)
        self.connection.commit()
        print(f"Data has been deleted from {table_name} table.")

    def close(self):
        self.cursor.close()
        self.connection.close()

# Usage example
db = Database("localhost","root", "Qwerty123", "my_database")
db.create_table("employees", "id INT(6) PRIMARY KEY, name VARCHAR(30), age INT(3)")
db.insert_data("employees", "1, 'John Smith', 25")
db.select_data("employees")
db.update_data("employees", "age", 30, "id = 1")
db.select_data("employees")
db.delete_data("employees", "id = 1")
db.select_data("employees")
db.close()

```

```python
import mysql.connector

class Stock:
    def __init__(self, vendor_name, stock_name, stock_quantity, stock_price, location):
        self.vendor_name = vendor_name
        self.stock_name = stock_name
        self.stock_quantity = stock_quantity
        self.stock_price = stock_price
        self.location = location

class StockManagement:
    def __init__(self):
        self.cnx = mysql.connector.connect(
            host="localhost",
            user="root",
            password="Qwerty123"
        )
        self.cursor = self.cnx.cursor()

    def create_database(self):
        self.cursor.execute("CREATE DATABASE IF NOT EXISTS stock_management")
        self.cursor.execute("use stock_management")
        print("Database created successfully.")

    def create_tables(self):
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS vendors (
                vendor_id INT AUTO_INCREMENT PRIMARY KEY,
                vendor_name VARCHAR(255) NOT NULL
            )""")
        self.cursor.execute("ALTER TABLE vendors ADD UNIQUE(vendor_name)")
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS stocks (
                stock_id INT AUTO_INCREMENT PRIMARY KEY,
                stock_name VARCHAR(255) NOT NULL,
                stock_quantity INT NOT NULL,
                stock_price DECIMAL(10,2) NOT NULL,
                location VARCHAR(255) NOT NULL,
                vendor_id INT,
                FOREIGN KEY (vendor_id) REFERENCES vendors(vendor_id)
            )""")
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS transactions (
                transaction_id INT AUTO_INCREMENT PRIMARY KEY,
                stock_id INT,
                transaction_type ENUM('buy', 'sell'),
                transaction_quantity INT NOT NULL,
                transaction_date DATE NOT NULL,
                FOREIGN KEY (stock_id) REFERENCES stocks(stock_id)
            )""")
        print("Tables created successfully.")


    def add_vendor(self, vendor_name):
        
        add_vendor = ("INSERT INTO vendors (vendor_name) VALUES (%s)")
        vendor_data = (vendor_name,)
        self.cursor.execute(add_vendor, vendor_data)
        self.cnx.commit()
        print("Vendor added successfully.")

    def add_stock(self, stock):
        add_stock = ("INSERT INTO stocks (stock_name, stock_quantity, stock_price, location, vendor_id) VALUES (%s, %s, %s, %s, (SELECT vendor_id FROM vendors WHERE vendor_name = %s))")
        stock_data = (stock.stock_name, stock.stock_quantity, stock.stock_price, stock.location, stock.vendor_name)
        self.cursor.execute(add_stock, stock_data)
        self.cnx.commit()
        print("Stock added successfully.")

    def buy_stock(self, stock_id, quantity):
        update_stock = ("UPDATE stocks SET stock_quantity = stock_quantity + %s WHERE stock_id = %s")
        stock_data = (quantity, stock_id)
        self.cursor.execute(update_stock, stock_data)
        self.cnx.commit()
        add_transaction = ("INSERT INTO transactions (stock_id, transaction_type, transaction_quantity, transaction_date) VALUES (%s, 'buy', %s, CURRENT_DATE())")
        transaction_data = (stock_id, quantity)
        self.cursor.execute(add_transaction, transaction_data)
        self.cnx.commit()
        print("Stock purchased successfully.")

    def sell_stock(self, stock_id, quantity):
        check_stock = ("SELECT stock_quantity FROM stocks WHERE stock_id = %s")
        check_data = (stock_id,)
        self.cursor.execute(check_stock, check_data)
        result = self.cursor.fetchone()
        if result[0] < quantity:
            print("Sorry, we do not have enough stock to sell.")
        else:
            update_stock = ("UPDATE stocks SET stock_quantity = stock_quantity - %s WHERE stock_id = %s")
            stock_data = (quantity, stock_id)
            self.cursor.execute(update_stock, stock_data)
            self.cnx.commit()
            add_transaction = ("INSERT INTO transactions (stock_id, transaction_type, transaction_quantity, transaction_date) VALUES (%s, 'sell', %s, CURRENT_DATE())")
            transaction_data = (stock_id, quantity)
            self.cursor.execute(add_transaction, transaction_data)
            self.cnx.commit()
            print("Stock sold successfully.")

    def check_stock_quantity(self, stock_id):
        check_stock = ("SELECT stock_quantity FROM stocks WHERE stock_id = %s")
        check_data = (stock_id,)
        self.cursor.execute(check_stock, check_data)
        result = self.cursor.fetchone()
        print("Stock quantity: ", result[0])

    def check_vendor_stocks(self, vendor_name):
        check_vendor = ("SELECT stock_name, stock_quantity, stock_price, location FROM stocks WHERE vendor_id = (SELECT vendor_id FROM vendors WHERE vendor_name = %s)")
        vendor_data = (vendor_name,)
        self.cursor.execute(check_vendor, vendor_data)
        result = self.cursor.fetchall()
        print("Vendor Stocks:")
        for row in result:
            print("Stock Name: ", row[0])
            print("Stock Quantity: ", row[1])
            print("Stock Price: ", row[2])
            print("Stock Location: ", row[3])
            print()

    def check_transactions(self):
        check_transactions = ("SELECT stock_name, transaction_type, transaction_quantity, transaction_date FROM transactions JOIN stocks ON transactions.stock_id = stocks.stock_id")
        self.cursor.execute(check_transactions)
        result = self.cursor.fetchall()
        print("Transactions:")
        for row in result:
            print("Stock Name: ", row[0])
            print("Transaction Type: ", row[1])
            print("Transaction Quantity: ", row[2])
            print("Transaction Date: ", row[3])
            print()

    def close_connection(self):
        self.cursor.close()
        self.cnx.close()
        print("Connection closed.")
```

```python
# Create a new instance of the StockManagement class
stock_manager = StockManagement()

# Create the database and tables
stock_manager.create_database()
stock_manager.create_tables()

# Add some vendors
stock_manager.add_vendor("Vendor1")
stock_manager.add_vendor("Vendor")
```

```python

# Add some vendors
stock_manager.add_vendor("Vendor 1")
stock_manager.add_vendor("Vendor 2")

# Add some stocks
stock1 = Stock("Vendor 1", "Stock 1", 50, 10.50, "New York")
stock2 = Stock("Vendor 2", "Stock 2", 30, 12.00, "Los Angeles")
stock3 = Stock("Vendor 1", "Stock 3", 40, 15.25, "Chicago")
stock_manager.add_stock(stock1)
stock_manager.add_stock(stock2)
stock_manager.add_stock(stock3)

# Buy some stocks
stock_manager.buy_stock(1, 10)
stock_manager.buy_stock(2, 20)

# Sell some stocks
stock_manager.sell_stock(1, 5)
stock_manager.sell_stock(2, 25)

# Check stock quantities
stock_manager.check_stock_quantity(1)
stock_manager.check_stock_quantity(2)

# Check vendor stocks
stock_manager.check_vendor_stocks("Vendor 1")

# Check transactions
stock_manager.check_transactions()

# Close the connection
stock_manager.close_connection()
```
